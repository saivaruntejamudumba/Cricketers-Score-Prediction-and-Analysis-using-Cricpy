```python
X=main_df.drop(['Runs'],axis=1)
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc=LabelEncoder()
```

```python
X['Opposition']=enc.fit_transform(X['Opposition'])
X['Dismissal']=enc.fit_transform(X['Dismissal'])
```

```python
X.head(20)
```

```python
opp=pd.get_dummies(X['Opposition'])
```

```python
opp.head(20)
```

```python
dis=pd.get_dummies(X['Dismissal'])
dis.head()
```

```python
opp=opp.drop([0],axis=1)
```

```python
opp.head()
```

```python
dis=dis.drop([0],axis=1)
dis.head()
```

```python
dis.columns=['d1','d2','d3','d4','d5']
```

```python
dis.head()
```

```python
dum=pd.concat([dis,opp],axis=1)
```

```python
dum.head()
```

```python
X=X.drop(['Opposition','Dismissal'],axis=1)
X.head()
```

```python
X=pd.concat([X, dum],axis=1)
X.head()
```

```python
y=main_df['Runs']
len(y)
```

### Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
model=reg.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error
```

```python
import math
r=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=math.sqrt(mse)
print(" R2 value:",r, "\n")
print(" MSE value:",mse,"\n")
print(" RMSE value:",rmse,"\n")
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
xgb=GradientBoostingRegressor()
```

```python
model2=xgb.fit(X_train,y_train)
```

```python
y_pred=model2.predict(X_test)
```

```python
r=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=math.sqrt(mse)
print(" R2 value:",r, "\n")
print(" MSE value:",mse,"\n")
print(" RMSE value:",rmse,"\n")
```
